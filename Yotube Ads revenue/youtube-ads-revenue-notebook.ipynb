# <h1>Importing Required Libraries </h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h1>Loading Dataser </h1>

```python
df = pd.read_csv('/kaggle/input/youtube-ads-revenue/youtube ads.csv')
```

```python
#Checking Datatype
df.dtypes
```

```python
#Converting 'Date' column from object to datetime type
df['Date'] = df['Date'].astype('datetime64')
```

<h1>Exploring data </h1>

```python
#adding year column
df['year'] = df['Date'].dt.year
```

```python
#Grouping data by year
df_year = df.groupby('year')
```

```python
#Calculating Mean revenue, total revenue each year
df_mean_revenue = df_year.mean()
print(df_mean_revenue)
```

```python
df_total_revenue = df_year.sum()
df_total_revenue
```

<h1> Plotting Ads Revenue </h1>

```python
df.plot(x='Date',y ='Ads Revenue (Mn)',kind='line',title='Youtube : Ads Revenue')
plt.xlabel('Year of Date')
plt.ylabel('Revenue in Ten Billions')
plt.grid(True)
plt.show()
```
