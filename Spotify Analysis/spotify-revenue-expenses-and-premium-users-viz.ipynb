<h1>Importing Required Libraries</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

<h1>Loading Dataset</h1>

```python
df = pd.read_csv('/kaggle/input/spotify-revenue-expenses-and-its-premium-users/Spotify Quarterly.csv')
```

*Since, last row is incomplete, so we will not consider last row in our analysis*

```python
df = df[0:len(df)-1]
```

<h1>Formatting Dataset</h1>
* Checking data types

```python
df.dtypes
```

We found that all columns have required data types except date which is objcet data type.

```python
#Converting "Date" into date data type
df['Date']= df['Date'].astype('datetime64')
```

<h1>Visualizng Dataset<h1>

<h3>Plotting Total revenue and Cost of Revenue</h3>

```python
plt.plot('Date','Total Revenue', data=df,color='g',label='Total Revenue')
plt.plot('Date','Cost of Revenue', data=df,color='r',label='Cost of Revenue')
plt.xlabel('Year(Date)')
plt.ylabel('In Euro Millions')
plt.grid(True)
plt.title('Spotify Total Revenue & Expenses Over Time')
plt.tight_layout()
plt.legend()
plt.show()
```

<h4>Conclusion</h4>

Over the year, Spotify's total revenue increased as well as difference in expenses and revenue is alos increased. So, it means Gross profit increased a lot. 


<h3>Plotting Total revenue and Gross profit </h3>


```python
plt.plot('Date','Total Revenue', data=df,color='g',label='Total Revenue')
plt.plot('Date','Gross Profit', data=df,color='r',label='Gross Profit')
plt.xlabel('Year(Date)')
plt.ylabel('In Euro Millions')
plt.grid(True)
plt.title('Spotify Total Revenue & Gross Profit Over Time')
plt.tight_layout()
plt.legend()
plt.show()
```

<h3>Comparing Premium Revenue and Ad Revenue</h3>

```python
plt.plot('Date','Premium Revenue', data=df,color='g',label='Premium Users Revenue')
plt.plot('Date','Ad Revenue', data=df,color='r',label='Ad Supported Users Revenue')
plt.xlabel('Year(Date)')
plt.ylabel('In Euro Millions')
plt.grid(True)
plt.title('Premium Users & Ad Users Revenue Over Time')
plt.tight_layout()
plt.legend()
plt.show()
```

<h3>Plotting Total Users</h3>

```python
plt.plot('Date','MAUs', data=df)
plt.xlabel('Year(Date)')
plt.ylabel('Users In Millions')
plt.grid(True)
plt.title('Spotify : Monthly Active Users')
plt.tight_layout()
plt.show()
```

<h3> Comparing Premium Users and Ad Supported Users </h3>

```python
plt.plot('Date','Premium MAUs', data=df,color='g',label='Premium Users')
plt.plot('Date','Ad MAUs', data=df,color='r',label='Ad Supported Users')
plt.xlabel('Year(Date)')
plt.ylabel('Users In Millions')
plt.grid(True)
plt.title('Spotify : Premium Users Vs Ad-Supported Users')
plt.tight_layout()
plt.legend()
plt.show()
```

<h4>Conclusion</h4>
Over the time, difference in Ad-Supported Users and Premium Users increased, i.e, there are more Ad-Supported Users as always.

Even though, Number of premium users are less,the premium revnue is too high in comparison to Ad-Supported Users. But, is it smae case in Gross Profit. Let check it.



<h3> Spotify Gross Profit From : Premium Users Vs Ad-Supported Users </h3>

```python
plt.plot('Date','Premium Gross Profit', data=df,color='g',label='Premium Gross Profit')
plt.plot('Date','Ad gross Profit', data=df,color='r',label='Ad Supported Gross Profit')
plt.xlabel('Year(Date)')
plt.ylabel('Users In Millions')
plt.grid(True)
plt.title('Spotify Gross Profit From : Premium Users Vs Ad-Supported Users')
plt.tight_layout()
plt.legend()
plt.show()
```

<h4>Conclusion:</h4>
Spotify Gross Profit generated from Premium users is too high than Ad-Supported Users, even though Ad Users are far higher in number.

<h3> Spotify's Premium Users : Average Revenue Per Users </h3>

```python
plt.plot('Date','Premium ARPU', data=df)
plt.xlabel('Year(Date)')
plt.ylabel('In Euro')
plt.grid(True)
plt.title('Spotify Premium Users : Average Revenue Per Users')
plt.tight_layout()
plt.show()
```

<h4>Conclusion</h4>
<p style=color:red;>Spotify Premium Revenue per users have fallen drastically over the years.</p>
